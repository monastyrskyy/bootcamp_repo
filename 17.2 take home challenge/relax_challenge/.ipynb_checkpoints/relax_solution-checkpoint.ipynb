```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import warnings
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import sklearn
from datetime import timedelta
from sklearn.metrics import roc_curve
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

```

```python
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 25)
```

```python
data_users = pd.read_csv('takehome_users.csv', encoding='latin-1')
```

```python
data_users
```

```python
data_eng = pd.read_csv('takehome_user_engagement.csv')
data_eng.time_stamp = pd.to_datetime(data_eng.time_stamp)
data_eng = data_eng.set_index('time_stamp', drop = True)
```

```python
data_eng#.groupby('user_id').sum().head(20)
```

```python
# help from various sources online

def adopted_func(x):
    label = 0
    data = data_eng[data_eng['user_id'] == x].resample('D').min().dropna()
    for i in range(len(data)-2):
        if data.index[i + 2] - data.index[i] > timedelta(days=7):
            label = 0
        else:
            label = 1
    return label

data_users['adopted'] = data_users['object_id'].apply(adopted_func)
```

```python
data_users.groupby('adopted').count()
```

```python
# removing all unnecessary columns
# Removing a few to make sure that small cell sizes don't exist
data_users.drop(['object_id', 'creation_time', 'name', 'email', 'invited_by_user_id', 'last_session_creation_time'
                 ], axis = 1, inplace = True)
```

```python
# Setting up dummies
# One hot encoding
model_data = pd.get_dummies(data_users, columns = ['creation_source', 'org_id'
                                                   ], drop_first = True)
```

```python
for col in model_data.columns:
    if model_data[col].isnull().values.sum() > 0:
        print(col, ":", model_data[col].isnull().values.sum(), 'missing values.')
```

```python
# Initial split
y = model_data.loc[:,'adopted']
x = model_data.drop('adopted', axis = 1)
```

```python
# Split into training and test set
# Stratifying to make sure the split of phisihng/valid stays equal across train/test split.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=2)
```

```python

rf = RandomForestClassifier(random_state=2)

params_rf = {
             'n_estimators': [100, 350, 500],
             'min_samples_leaf': [2, 10, 30]
             }

Randomized_rf = GridSearchCV(estimator=rf,param_grid=params_rf,
                     scoring='accuracy',cv=5)

Randomized_rf.fit(x_train, y_train)
```

```python
# approach help from various sources online and sdf94 on github
best_rf = RandomForestClassifier(bootstrap=True, class_weight=None,
                                              criterion='gini', max_depth=None,
                                              max_features='auto',
                                              max_leaf_nodes=None,
                                              min_impurity_decrease=0.0,
                                              min_impurity_split=None,
                                              min_samples_leaf=1,
                                              min_samples_split=2,
                                              min_weight_fraction_leaf=0.0,
                                              n_estimators='warn', n_jobs=None,
                                              oob_score=False, random_state=2,
                                              verbose=0, warm_start=False)
best_rf.fit(x_train,y_train)
feature_coef = pd.DataFrame(best_rf.feature_importances_).transpose()
feature_coef.columns = list(x_train.columns)
feature_coef.index = ['RF'] 
feature_coef.transpose().sort_values(by='RF',ascending=False).head(20)
```

Above are the top 20 features that affect people becoming adopted users. I ran a Random Forest classifier because of the time contraint of the assignment as well as RF being a good method. It's possible that there are other methods that would produce better results, but this is at the very least, a good foundation that can easily deal with very wide datasets, which given the encoding on this dataset, it is.

From the results of my analysis, it looks like those users who are open to being in a mailing list and open to receiving marketing material are the users that will likely be considered adopted. Also those who signed up as a result of being in an organization is another good predictor of them being adopted. Some interesting organizations that seem to have a high influence are org_id_387, and org_id_318, among others.

SOme other factors that I considered in my investigation were the null values, which columns were actually useful for this analysis, prompting me to remove several columns that would not have been useful. Luckily, there weren't missing values in the columns that I found useful.

As for future steps, it would be useful to know how much each person uses the app for, their geographical location, and perhaps other demographic data such as age, gender,etc. which can all play a role in deciding the accuracy of the model.
